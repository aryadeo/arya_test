```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
#There's a convenient way for plotting objects with labelled data (i.e.data that can be accessed by index ``obj['y']``). 
#Instead of giving bthe data in *x* and *y*, you can provide the object in the *data* parameter and just give the labels for *x* and *y*::
#plot('xlabel', 'ylabel', data=obj)
plt.scatter('a', 'b',s='c',c='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

#Basics of pyplot

```python
#example of auto indexing
plt.plot([1,2,3,4,5,6,7])
plt.show()
```

```python
x=np.arange(10)
y=np.linspace(0,1,10)
plt.plot(x,y,'ro')
plt.title('sample plot')
plt.xlabel('index')
plt.ylabel('value')
plt.show()
```

```python
#multiple plot from a single command
x=np.linspace(0,10,360)
y=np.cos(2*np.pi*1*x)
y2=np.sin(2*np.pi*1*x)
plt.plot(x,y,'ro',x,y2,'b--')
plt.show()
```

```python
#bar plot
x=['category_1','category_2','category_3']
y=np.arange(3)
plt.bar(x,y)#size of input araguments should match

```

```python
#Categorical figure(inspired from default tutorial)
plt.figure(figsize=(20,8))#maintaning a fixed figure size just to display good
x=['category_1','category_2','category_3']
y=np.random.random((1,3))
plt.subplot(131)#looking for 1st column of 1 X 3 plot matrix
plt.bar(x,y[0])#as y is of dimension 2 we are using the 1st index of y to plot against a category

plt.subplot(132)
plt.scatter(x,y[0])

plt.subplot(133)
plt.plot(x,y[0])

plt.show()

```

```python
#multiple figures
plt.figure(num=1)
x=np.linspace(0,10,360)
y=np.cos(2*np.pi*1*x)
plt.plot(x,y,'ro',)

plt.figure(num=2)
y1=np.sin(2*np.pi*1*x)
plt.plot(x,y1,'bo',)

plt.show()
#here we are creating two different figures
```

```python
#1D histogram. 
#There are four type of gistogram. the bar tpe, stacked basr type, step type(normal unfilled line plot) and step filled type.
#The hist funtion returns a tuple of (n, bins, patches)
mu, sigma=100,15
x=mu+sigma*np.random.randn(1000)
plt.figure(num=1)
n=plt.hist(x)
plt.show()

#now change the parameters and see the effect
#just uncomment each line and see the effectes of each. Change the parameters and observe
#plt.figure(num=2)
#plt.hist(x, bins=np.arange(200))
#plt.hist(x,range=(1,200),color='brown')
#plt.hist(x, orientation='vertical',color='blue',histtype='step',align='right')#change the alignment to left and mid
#plt.show()
```

```python
#text and annotations
#TeX expression can be written as surrounded by dollar signs
plt.figure()
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line = plt.plot(t, s, lw=4)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),arrowprops=dict(width=4, headwidth=10
                                                                     ,facecolor='red', shrink=0.05))

plt.ylim(-2, 2)
plt.title(r'$\sigma_i=15$')
plt.text( 1,1.8,'$\mu=100,\ \sigma_i=15$')
plt.show()
```
